```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python

with open('gasDemo.csv', newline='') as f:
    reader = csv.reader(f)
    column = [row[1] for row in reader]

a1 = pd.read_csv('gasDemo.csv', usecols=['XX进口压力'],encoding='gbk')
ax1 = plt.subplot(2, 3, 1)
plt.plot(a1, color='k')

a2 = pd.read_csv('gasDemo.csv', usecols=['XX出口压力'],encoding='gbk')
ax2 = plt.subplot(2, 3, 2)
plt.plot(a2, color='k')

a3 = pd.read_csv('gasDemo.csv', usecols=['XX过滤器压差1号'],encoding='gbk')
ax3 = plt.subplot(2, 3, 3)
plt.plot(a3, color='k')

a4 = pd.read_csv('gasDemo.csv', usecols=['XX过滤器压差2号'],encoding='gbk')
ax4 = plt.subplot(2, 3, 4)
plt.plot(a4, color='k')


a5 = pd.read_csv('gasDemo.csv', usecols=['XX2号流量计工况瞬时流量'],encoding='gbk')
ax5 = plt.subplot(2, 3, 5)
plt.plot(a5, color='k')

a6 = pd.read_csv('gasDemo.csv', usecols=['XX2号流量计标况瞬时流量'],encoding='gbk')
ax6 = plt.subplot(2, 3, 6)
plt.plot(a6, color='k')
```
