```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import csv
```

```python

# with open('gasDemo.csv', newline='') as f:
#     reader = csv.reader(f)
#     column = [row[1] for row in reader]

plt.figure(figsize=(20,10))

a1 = pd.read_csv('gasDemo.csv', usecols=['XX进口压力'],encoding='gbk')
ax2 = plt.subplot(2, 3, 1)
plt.plot(a1, color='k')
plt.title('inlet pressure')

a2 = pd.read_csv('gasDemo.csv', usecols=['XX出口压力'],encoding='gbk')
ax2 = plt.subplot(2, 3, 2)
plt.plot(a2, color='k')
plt.title('outlet pressure')

a3 = pd.read_csv('gasDemo.csv', usecols=['XX过滤器压差1号'],encoding='gbk')
ax3 = plt.subplot(2, 3, 3)
plt.plot(a3, color='k')
plt.title('pressure difference 1')

a4 = pd.read_csv('gasDemo.csv', usecols=['XX过滤器压差2号'],encoding='gbk')
ax4 = plt.subplot(2, 3, 4)
plt.plot(a4, color='k')
plt.title('pressure difference 2')


a5 = pd.read_csv('gasDemo.csv', usecols=['XX2号流量计工况瞬时流量'],encoding='gbk')
ax5 = plt.subplot(2, 3, 5)
plt.plot(a5, color='k')
plt.title('operating case flow')

a6 = pd.read_csv('gasDemo.csv', usecols=['XX2号流量计标况瞬时流量'],encoding='gbk')
ax6 = plt.subplot(2, 3, 6)
plt.plot(a6, color='k')
plt.title('standard flow')

plt.legend(loc='upper right')
```

```python
a1 = pd.read_csv('gasDemo.csv', usecols=['XX进口压力'],encoding='gbk')
plt.plot(a1, color='b', label='inlet pressure')

a2 = pd.read_csv('gasDemo.csv', usecols=['XX出口压力'],encoding='gbk')
plt.plot(a2, color='r', label='outlet pressure')

plt.legend(loc='upper right')
plt.ylim(0, 300)
```

```python
a3 = pd.read_csv('gasDemo.csv', usecols=['XX过滤器压差1号'],encoding='gbk')

plt.plot(a3, color='b', label='Pressure difference 1')
plt.legend(loc='upper right')
plt.show()
#
a4 = pd.read_csv('gasDemo.csv', usecols=['XX过滤器压差2号'],encoding='gbk')

plt.plot(a4, color='r', label='Pressure difference 2')

plt.legend(loc='upper right')
plt.show()

plt.plot(a3, color='b', label='Pressure difference 1')
plt.plot(a4, color='r', label='Pressure difference 2')
plt.legend(loc='upper right')
# plt.ylim(0, 300)
```

```python
a4 = pd.read_csv('gasDemo.csv', usecols=['XX2号流量计工况瞬时流量'],encoding='gbk')

plt.plot(a4, color='b', label='operating case flow')
plt.legend(loc='upper right')
plt.show()
#
a5 = pd.read_csv('gasDemo.csv', usecols=['XX2号流量计标况瞬时流量'],encoding='gbk')

plt.plot(a5, color='r', label='standard flow')

plt.legend(loc='upper right')
plt.show()

plt.plot(a4, color='b', label='operating case flow')
plt.plot(a5, color='r', label='standard flow')
plt.legend(loc='upper right')
```
